```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(start= -3, stop= 3, num=100)
```

```python
# Cost Function 
def f(x):
    return x**2 + x + 1

# Slope of the cost function (Derivative)
def df(x):
    return 2*x + 1
```

```python
plt.figure(figsize=[20, 10])

plt.subplot(1, 2, 1)
plt.plot(x, f(x))
plt.title("Cost Function")
plt.xlabel("Data")
plt.ylabel("f(Data)")


plt.subplot(1, 2, 2)
plt.plot(x, df(x))
plt.title("Cost Function")
plt.xlabel("Data")
plt.ylabel("df(Data)")

plt.show()
```

```python
# Gradient Decent

count = 100
new_x = 3
prev_x = 0
step_multi = 0.1
precision = 0.0001
x_new = [new_x]

for i in range(100):
    prev_x = new_x
    gradient = df(prev_x)
    new_x = prev_x - (step_multi * gradient)
    x_new.append(new_x)
    step = abs(new_x - prev_x)
    if step < precision:
        break

x_new = np.array(x_new)
print(f"Local Minimum : {new_x}")
print(f"Cost at the local minimum ({new_x}) is : {f(new_x)}")
print(f"Slop of the local minimum ({new_x}) is : {df(new_x)}")
```

```python
plt.figure(figsize=[20, 10])

plt.subplot(1, 2, 1)
plt.scatter(x_new, f(x_new))
plt.plot(x, f(x))
plt.title("Cost Function")
plt.xlabel("Data")
plt.ylabel("f(Data)")
plt.xlim(-2, 2)
plt.ylim(-0.5, 2)


plt.subplot(1, 2, 2)
plt.scatter(x_new, df(x_new))
plt.plot(x, df(x))
plt.title("Cost Function")
plt.xlabel("Data")
plt.ylabel("df(Data)")
plt.xlim(-2, 2)
plt.ylim(-0.5, 2)

plt.show()
```

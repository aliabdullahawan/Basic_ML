```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(start= -2, stop= 2, num= 100)
```

```python
def g(x):
    return x**4 - 4*x**2 + 5

def dg(x):
    return 4*x**3 - 8*x
```

```python
plt.figure(figsize=[20, 10])

plt.subplot(1, 2, 1)
plt.plot(x, g(x))
plt.title("Cost Function")
plt.xlabel("Data")
plt.ylabel("f(Data)")
plt.xlim(-2, 2)
plt.ylim(0, 6)
plt.grid()

plt.subplot(1, 2, 2)
plt.plot(x, dg(x))
plt.title("Cost Function")
plt.xlabel("Data")
plt.ylabel("df(Data)")
plt.xlim(-2, 2)
plt.ylim(-6, 8)
plt.grid()

plt.show()
```

```python
def Gradiend_Decent(derivative_func, initial_guess, learning_rate= 0.01, precision= 0.001, total_run= 100):
    count = total_run
    new_x = initial_guess
    x_new = [new_x]
    
    for i in range(count):
        prev_x = new_x
        gradient = derivative_func(new_x)
        new_x = prev_x - (gradient * learning_rate)
        x_new.append(new_x)
        step = abs(new_x - prev_x)
        if step < precision:
            break
    return new_x, x_new
```

```python
new_x, x_new =  Gradiend_Decent(derivative_func= dg, initial_guess= 0.1)
```

```python
plt.figure(figsize=[20, 10])

plt.subplot(1, 2, 1)
plt.scatter(x_new, g(np.array(x_new)))
plt.plot(x, g(x))
plt.title("Cost Function")
plt.xlabel("Data")
plt.ylabel("f(Data)")
plt.xlim(-2, 2)
plt.ylim(0, 6)
plt.grid()

plt.subplot(1, 2, 2)
plt.scatter(x_new, dg(np.array(x_new)))
plt.plot(x, dg(x))
plt.title("Cost Function")
plt.xlabel("Data")
plt.ylabel("df(Data)")
plt.xlim(-2, 2)
plt.ylim(-6, 8)
plt.grid()

plt.show()
```

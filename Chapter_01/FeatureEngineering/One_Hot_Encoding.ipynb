# ONE HOT ENCODING

### Using Pandas

Without droping (n)

```python
p_ohe = pd.get_dummies(data= cars, columns= ["fuel", "owner"])
p_ohe
```

K-1 OHE (With droping (n-1))

```python
k1_ohe = pd.get_dummies(data= cars, columns= ["fuel", "owner"], drop_first= True)
k1_ohe
```

### Using Scikit-Learn

```python
x_train, x_test, y_train, y_test = skl.model_selection.train_test_split(cars.iloc[::, :4:], cars.iloc[::, -1], test_size= 0.2, random_state= 0)
oneHotEncoding = skl.preprocessing.OneHotEncoder(drop= "first", sparse_output= True, dtype= np.int32)
```

```python
# One Hot Encoder always produce Sparse Matrix
x_train_ohe = oneHotEncoding.fit_transform(x_train[["fuel", "owner"]]).toarray()
x_test_ohe = oneHotEncoding.transform(x_test[["fuel", "owner"]]).toarray()
```

```python
aa = np.hstack((x_train[["brand", "km_driven"]].values, x_train_ohe)) # Remember use the tuple here in hstack function ()
```

```python
pd.DataFrame(aa)
```

## ONE HOT ENCODING FOR BRAND (Which has many catagories)

```python
count = cars["brand"].value_counts()
count
```

```python
cars["brand"].nunique()
threashold = 100
```

```python
repl = count[count <= threashold].index
repl
```

```python
mv_ohe = pd.get_dummies(cars["brand"].replace(repl, "uncommon"))
mv_ohe
```

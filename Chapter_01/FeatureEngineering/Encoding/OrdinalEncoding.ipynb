```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn as skl
```

```python
customer = pd.read_csv("./customer.csv")
customer.sample(10)
```

```python
customer = customer.iloc[::, 2::]
customer.sample(10)
```

```python
x_train, x_test, y_train, y_test = skl.model_selection.train_test_split(customer.iloc[::, :2:], customer.iloc[::, -1], test_size= 0.2)


ordinalEncoder = skl.preprocessing.OrdinalEncoder(categories=[["Poor", "Average", "Good"], ["PG", "School", "UG"]])

ordinalEncoder.fit(x_train)
x_train_oe = ordinalEncoder.transform(x_train)
x_test_oe = ordinalEncoder.transform(x_test)

x_train_oe = pd.DataFrame(x_train_oe, columns= x_train.columns, dtype= int)
x_test_oe = pd.DataFrame(x_test_oe, columns= x_test.columns, dtype= int)
```

```python
x_test_oe.sample(10)
```

```python
ordinalEncoder.categories_
```

```python
labelEncoding = skl.preprocessing.LabelEncoder()

labelEncoding.fit(y_train)
y_train_le = labelEncoding.transform(y_train)
y_test_le = labelEncoding.transform(y_test)

y_train_le = pd.DataFrame(y_train_le)
y_test_le = pd.DataFrame(y_test_le)
```

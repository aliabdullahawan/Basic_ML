```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn as skl
import scipy as scp
```

```python
concrete = pd.read_csv("./concrete_data.csv")
concrete.sample(10)
```

```python
concrete.isnull().sum()
```

```python
concrete.describe()
```

```python
x = concrete.iloc[::, :8:]
y = concrete.iloc[::, -1]
```

```python
x_train, x_test, y_train, y_test = skl.model_selection.train_test_split(x, y, test_size= 0.2, random_state=42)
```

```python
linearRegression = skl.linear_model.LinearRegression()

linearRegression.fit(x_train, y_train)

y_predicion = linearRegression.predict(x_test)

skl.metrics.r2_score(y_test, y_predicion)
```

```python
linearRegression = skl.linear_model.LinearRegression()
np.mean(skl.model_selection.cross_val_score(linearRegression, x, y, scoring= "r2"))
```

```python
for col in x_train.columns:
    plt.figure(figsize=(14, 5))

    plt.subplot(121)
    sns.distplot(x_train[col])
    plt.title(col)
    
    plt.subplot(122)
    scp.stats.probplot(x_train[col], dist= "norm", plot= plt)
    plt.title(col)
    
    plt.show()
```

```python
pt_bc = skl.preprocessing.PowerTransformer(method= "box-cox")

x_train_pt_bc = pt_bc.fit_transform(x_train + 0.000001)
x_test_pt_bc = pt_bc.transform(x_test + 0.000001)

lambdas_pt_bc = pd.DataFrame({"cols": x_train.columns, "box_cox_lambdas": pt_bc.lambdas_})
lambdas_pt_bc
```

```python
linearRegression = skl.linear_model.LinearRegression()

linearRegression.fit(x_train_pt_bc, y_train)

y2_pred = linearRegression.predict(x_test_pt_bc)

skl.metrics.r2_score(y_test, y2_pred)
```

```python
x_train_pt_bc = pd.DataFrame(data= x_train_pt_bc, columns= x_train.columns)
x_train_pt_bc
```

```python
for col in x_train_pt_bc.columns:
    plt.figure(figsize=(20, 4))

    plt.subplot(141)
    sns.distplot(x_train[col])
    plt.title(col)
    
    plt.subplot(142)
    scp.stats.probplot(x_train[col], dist= "norm", plot= plt)
    plt.title(col)
    
    plt.subplot(143)
    sns.distplot(x_train_pt_bc[col])
    plt.title(col)
    
    plt.subplot(144)
    scp.stats.probplot(x_train_pt_bc[col], dist= "norm", plot= plt)
    plt.title(col)
    
    plt.show()
```

```python
pt_yeo = skl.preprocessing.PowerTransformer()   # By default it uses Yeo-Johnson

x_train_pt_yeo = pt_yeo.fit_transform(x_train + 0.000001)
x_test_pt_yeo = pt_yeo.transform(x_test + 0.000001)

lambdas_pt_yeo = pd.DataFrame({"cols": x_train.columns, "box_cox_lambdas": pt_yeo.lambdas_})
lambdas_pt_yeo
```

```python
linearRegression = skl.linear_model.LinearRegression()

linearRegression.fit(x_train_pt_yeo, y_train)

y3_pred = linearRegression.predict(x_test_pt_yeo)

skl.metrics.r2_score(y_test, y3_pred)
```

```python
x_train_pt_yeo = pd.DataFrame(data= x_train_pt_yeo, columns= x_train.columns)
```

```python
for col in x_train_pt_yeo.columns:
    plt.figure(figsize=(20, 4))

    plt.subplot(141)
    sns.distplot(x_train[col])
    plt.title(col)
    
    plt.subplot(142)
    scp.stats.probplot(x_train[col], dist= "norm", plot= plt)
    plt.title(col)
    
    plt.subplot(143)
    sns.distplot(x_train_pt_yeo[col])
    plt.title(col)
    
    plt.subplot(144)
    scp.stats.probplot(x_train_pt_yeo[col], dist= "norm", plot= plt)
    plt.title(col)
    
    plt.show()
```

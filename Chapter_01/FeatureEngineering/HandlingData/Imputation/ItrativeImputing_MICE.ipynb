```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn as skl
from sklearn.experimental import enable_iterative_imputer
```

```python
startups = pd.read_csv("./50_Startups.csv")
startups.drop(columns= "State", inplace= True)

startups.iloc[1,0] = np.nan
startups.iloc[3,1] = np.nan
startups.iloc[-1,2] = np.nan

startups.sample(10)
```

```python
x = startups.iloc[::, :3:]
y = startups.iloc[::, -1]
```

```python
x_train, x_test, y_train, y_test = skl.model_selection.train_test_split(x, y, test_size= 0.2, random_state= 42)
```

```python
# x_train.iloc[1,0] = np.nan
# x_train.iloc[3,1] = np.nan
# x_train.iloc[-1,-1] = np.nan
```

```python
imp_itr = skl.impute.IterativeImputer(estimator= skl.linear_model.LinearRegression(), random_state= 0)

x_train_itr = imp_itr.fit_transform(x_train)
x_test_itr = imp_itr.transform(x_test)

```

```python
linearRegression = skl.linear_model.LinearRegression()
linearRegression.fit(x_train_itr, y_train)
y_pred = linearRegression.predict(x_test_itr)
```

```python
skl.metrics.r2_score(y_test, y_pred)
```

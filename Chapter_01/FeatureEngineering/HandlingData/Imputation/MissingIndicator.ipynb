```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn as skl
```

```python
titanic = pd.read_csv("./titanic_toy.csv")
titanic.sample(10)
```

```python
x = titanic.iloc[::, :3:]
y = titanic.iloc[::, -1]
```

```python
x_train, x_test, y_train, y_test = skl.model_selection.train_test_split(x, y, test_size= 0.2, random_state= 42)

mi = skl.impute.MissingIndicator()   # same as skl.imputer.SimpleImputer(add_indicator= True)
x_train_mi = mi.fit_transform(x_train)
x_test_mi = mi.transform(x_test)
```

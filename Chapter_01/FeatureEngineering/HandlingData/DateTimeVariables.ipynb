# HANDLING DATE

```python
# Extract Year, month, month name, day, day name, day of week, day of, week of year, quater
order["year"], order["month"], order["month_name"], order["week_of_year"], order["day"], order["day_name"], order["day_of_week"], order["day_of_year"], order["quater"] = order["date"].dt.year, order["date"].dt.month, order["date"].dt.month_name(), order["date"].dt.isocalendar().week, order["date"].dt.day, order["date"].dt.day_name(), order["date"].dt.dayofweek, order["date"].dt.dayofyear, order["date"].dt.quarter
order.sample(5)
```

```python
# Day is weekend or not
order["is_weekend"] = np.where(order["day_name"].isin(["Saturday", "Sunday"]), 1, 0)
order.sample(5)
```

```python
# What is the semester
order["semester"] = np.where(order["quater"].isin([1, 2]), 1, 2)
order.sample(5)
```

```python
import datetime
today = datetime.datetime.today()

today
```

```python
days = (today - order["date"]).dt.days
days
```

```python
# If you only need the whole number of months (floor)
months = (today.year - order["date"].dt.year) * 12 + (today.month - order["date"].dt.month)
months
```

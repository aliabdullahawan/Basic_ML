```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn as skl
```

```python
placement = pd.read_csv("./placement.csv")
placement.sample(10)
```

```python
plt.figure(figsize= (15, 5))

plt.subplot(121)
sns.distplot(placement["cgpa"])

plt.subplot(122)
sns.distplot(placement["placement_exam_marks"])
```

```python
placement["placement_exam_marks"].describe()
```

```python
sns.boxplot(x= placement["placement_exam_marks"])
plt.show()
```

```python
# Finding IQR
p25 = placement["placement_exam_marks"].quantile(0.25)
p75 = placement["placement_exam_marks"].quantile(0.75)

iqr = p75 - p25

# Finding bounds
lower_bound = p25 - (1.5 * iqr)
upper_bound = p75 + (1.5 * iqr)

print(
    f"P-25: {p25}\t"
    f"P-75: {p75}\t"
    f"IQR: {iqr}\t"
    f"Lower Bound: {lower_bound}\t"
    f"Upper Bound: {upper_bound}"
)
```

```python
# Finding outliers
placement[
    (placement["placement_exam_marks"] < lower_bound) |
    (placement["placement_exam_marks"] > upper_bound)
]
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Titanic-Dataset.csv")
df
```

# UNIVARIATE

## Catagorical Data

### Coun-Plot

```python
plt.figure(figsize=[20, 10])

plt.subplot(2, 2, 1)
plt.title("Survied Chart")
sns.countplot(df["Survived"])
df["Survived"].value_counts().plot(kind='bar')

plt.subplot(2, 2, 2)
plt.title("P-Class Type")
sns.countplot(df["Pclass"])
df["Pclass"].value_counts().plot(kind='bar')

plt.subplot(2, 2, 3)
plt.title("Sex")
df["Sex"].value_counts().plot(kind='bar')

plt.subplot(2, 2, 4)
plt.title("Embarked")
df["Embarked"].value_counts().plot(kind='bar')
```

```python
plt.figure(figsize=[13, 8])

plt.subplot(1, 4, 1)
plt.title("Survived")
df["Survived"].value_counts().plot(kind='pie', autopct="%.2f")

plt.subplot(1, 4, 2)
plt.title("P-Class Type")
df["Pclass"].value_counts().plot(kind='pie', autopct="%.2f")

plt.subplot(1, 4, 3)
plt.title("Sex")
df["Sex"].value_counts().plot(kind='pie', autopct="%.2f")

plt.subplot(1, 4, 4)
plt.title("Embarked")
df["Embarked"].value_counts().plot(kind='pie', autopct="%.2f")
```

# Numarical Data

### Histogram

```python
plt.hist(df["Age"], bins=15)
plt.grid()
```

### Dist-Plot

```python
# Kernel Density Estimation (KDE) 
# Probability Density Function (PDF) 
sns.histplot(df["Age"], kde= True)  
plt.grid()
```

```python
sns.distplot(df["Age"])
plt.grid()
```

### Box-Plot

```python
# minimum = Q1 - 1.5*IQR
# maximum = Q3 - 1.5*IQR
# IQR = Q3 - Q1
sns.boxplot(x= df["Age"])
```

```python
sns.boxplot(y= df["Fare"])
```

### Skewness

```python
# skew = 0 (Normal)
# skew = positive number (Right Skewed)
# skew = negative number (Let Skewed)
df["Age"].skew()
```

# MULTIVARIATE

### Numarical Numarical Combination (Scatter-Plot)

```python
sns.scatterplot(data= tips, x= "total_bill", y= "tip", hue= "sex", style= "smoker", size= "size")
```

### Numarical Catagorical Combination (bar-Plot)

```python
# The line is, Confidence Interval (CI) is a range of numbers (with a lower bound and an upper bound) 
# that we are fairly certain contains the true value we are trying to guess.
sns.barplot(data= titanic, x= "Pclass", y= "Age", hue= "Sex")
```

```python
sns.barplot(data= titanic, x= "Pclass", y= "Fare", hue= "Survived")
```

```python
sns.barplot(data= titanic, x= "Pclass", y= "Fare", hue= "Age")
```

### Numarical Catagorical Combination (Box-Plot)

```python
sns.boxplot(data= titanic, x= "Sex", y= "Age", hue= "Survived")
```

### Numarical Catagorical Combination (Dis-Plot)

```python
sns.distplot(titanic[titanic["Survived"]==0]["Age"])
sns.distplot(titanic[titanic["Survived"]==1]["Age"])
plt.legend(['Did Not Survive', 'Survived'])
```

```python
sns.histplot(titanic[titanic["Survived"]==0]["Age"], kde= True)
sns.histplot(titanic[titanic["Survived"]==1]["Age"], kde= True)
plt.legend(['Did Not Survive', 'Survived'])
```

```python
sns.distplot(titanic[titanic["Survived"]==0]["Age"], hist= False)
sns.distplot(titanic[titanic["Survived"]==1]["Age"], hist= False)
plt.legend(['Did Not Survive', 'Survived'])
```

### Catagorical Catagorical Combination (Heat-Map)

```python
print(pd.crosstab(titanic["Pclass"], titanic["Survived"]))
sns.heatmap(pd.crosstab(titanic["Pclass"], titanic["Survived"]))
```

```python
print(titanic.groupby("Pclass").mean(numeric_only= True)["Survived"] * 100)
(titanic.groupby("Pclass").mean(numeric_only= True)["Survived"] * 100).plot(kind= "bar")
```

```python
print(titanic.groupby("Sex").mean(numeric_only= True)["Survived"] * 100)
(titanic.groupby("Sex").mean(numeric_only= True)["Survived"] * 100).plot(kind= "bar")
```

```python
print(titanic.groupby("Embarked").mean(numeric_only= True)["Survived"] * 100)
(titanic.groupby("Embarked").mean(numeric_only= True)["Survived"] * 100).plot(kind= "bar")
```

### Catagorical Catagorical Combination (Cluster-Map)

```python
print(pd.crosstab(titanic["SibSp"], titanic["Survived"]))
sns.clustermap(pd.crosstab(titanic["SibSp"], titanic["Survived"]))
```
